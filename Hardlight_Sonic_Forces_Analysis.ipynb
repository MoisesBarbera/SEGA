#### To contribute a bit more, I have created a small table to observe the count of reviews for each rating provided. (It is easy to observe in a 100 review data set but, if this same code is used in a bigger data set with hundreds of thousands of reviews it will provide the corresponding values as well).

```python
five_star = df_sonic[df_sonic['rating'] == 5]  
four_star = df_sonic[df_sonic['rating'] == 4]
three_star = df_sonic[df_sonic['rating'] == 3]
two_star = df_sonic[df_sonic['rating'] == 2]
one_star = df_sonic[df_sonic['rating'] == 1]

df_ratings = pd.DataFrame({'rating' : ['5', '4', '3', '2', '1'],
                           'count' : [five_star['rating'].count(), four_star['rating'].count(),
                                      three_star['rating'].count(), two_star['rating'].count(),
                                      one_star['rating'].count()]})
df_ratings
```

#### With this information, I can also display on screen the exact percentage of reviews for each rating should we need a very exact value for a report or for a more accurate study.

Again, the code has been tailored to scalate and work with any dataset so that it can be incorporated in datasets with hundreds of thousands of reviews.

```python
print('5 Star Rating = {0} %'.format((five_star['rating'].count()/df_sonic['rating'].count())*100))
print('4 Star Rating = {0} %'.format((four_star['rating'].count()/df_sonic['rating'].count())*100))
print('3 Star Rating = {0} %'.format((three_star['rating'].count()/df_sonic['rating'].count())*100))
print('2 Star Rating = {0} %'.format((two_star['rating'].count()/df_sonic['rating'].count())*100))
print('1 Star Rating = {0} %'.format((one_star['rating'].count()/df_sonic['rating'].count())*100))
```

# 2 - Now that I have analysed a bit the numeric data it is time to analyse the qualitative data and apply Machine Learning and NLP to understand the verified reviews from all customers on the dataset.

```python
df_sonic['review'].iloc[:] # command to show on screen the reviews available on the dataset
```

```python
# To present it on a more elegant way, without displaying the ID


for i in df_sonic['review'].iloc[:5]:  # Displaying only the first 5 reviews to safe memory and space
    print(i, '\n')
```

```python
# Now I transform the dataframe structure into individual strings

words = df_sonic['review'].tolist() 
words[:5] # Displaying only the first 5 reviews to safe memory and space
```

```python
# And now all individual strings are combined together in one single string
# This means we are now selecting reviews as divided by characters (letters, punctuation signs and/or blank spaces)
#

words_as_one_string = ' '.join(words)
words_as_one_string[:500]  # Visualising first 500 characters as a demonstration of what this part of code does
```

```python
from wordcloud import WordCloud   # Needed to analyse words as a method for NLP
```

```python
plt.figure(figsize = (15,15))
plt.imshow(WordCloud().generate(words_as_one_string), interpolation = 'bilinear') #interpolation = 'bilinear' increases the quality of the image
plt.title("Overall words used on all reviews", fontsize=40)
plt.show()
```

#### As we can see, people think is a good game, fun to play, some relate the game to the movie.  On a smaller amount we find customers mentioning words like "suck", ad and problem; what means we should focus a lot on improving our ad display strategy and also focus on bringing the game to more tests and to identify the key problems discussed by users. 


#### Another big word used is Character, we should also investigate if it relates to good or bad reviews. Maybe they refer to how nice it is to play with their favourite characters or perhaps to how annoying it is that they have to pay to improve or obtain ccertain caharacters.


#### To obtain better insights, I am now going to study only the words from bad reviews and show how it compares with the above results to understand better where we should improve:

```python
df_bad_sonic = df_sonic[df_sonic['rating'] == 1]   #Creating a dataset with the bad reviews on it only (to know those reviews with 1 start).

bad_words = df_bad_sonic['review'].tolist()        #dataframe selected transformed into different strings
bad_words_as_one_string = ' '.join(bad_words)      #dataframe selected transformed into one single string
plt.figure(figsize = (15,15))
plt.imshow(WordCloud().generate(bad_words_as_one_string), interpolation = 'bilinear') #interpolation = 'bilinear' increases the quality of the image

plt.title("Most used words on negative reviews", fontsize=40)
plt.show()
```

#### As we can see from the bad reviews (which to recall are only an 8%), avoiding the key words that relate to the object under study (the "game") and the company ("SEGA"), the most repeated words are:

-  The Game Sucks
-  Pay
-  Need Change
-  Story Mode
-  Deleted
-  Hate
-  Crashing
-  Rethink
-  Characters

#### What suggests that a possible reason for the dislike from this 8% is that the need to pay often (or asking too often to pay for something, specially for characters after seen the data) makes users uncomfortable. Some users mentioned story mode, perhaps the single races and no chance of doing championshis nor haveing a bigger goal may lead to monotonous sessions. Some say that they deleted it but other beg the company to take action and change, that shows how even this customers who "hate" or "dislike" some aspects are still passionate about SEGA and the game and how they would be willing to play more if there were changes applied.
